# Proponiendo rutinas de limpieza que se pueden abordar para generar mayor precisión en los modelos y menor sesgo
en el análisis\.

## Valores Atípicos

## 1\. Tratamiento de Valores Nulos mediante Interpolación Temporal:

Para asegurar que los modelos predictivos \(como el pronóstico de cierres de puerto\) no hereden sesgos y sean lo más precisos posibles, se propone el siguiente pipeline de limpieza y preprocesamiento:

## 2\. Detección y Tratamiento de Valores Atípicos \(Outliers\):

  \- Justificación: Valores de viento o lluvia físicamente imposibles \(errores de sensor\) arrastrarán los pesos del modelo y reducirán la precisión\. Se abordará usando el Rango Intercuartílico \(IQR\) para detectar estos extremos y aplicar un recorte \(clipping\) en lugar de eliminarlos, ya que los temporales extremos \(aunque raros\) son claves para el negocio naviero\.

## 3\. Conversión y Extracción de Variables Temporales:

## 4\. Codificación de Variables Categóricas \(One\-Hot Encoding\):

 \- Justificación: Los datos meteorológicos son series de tiempo\. Rellenar nulos con la media global introduce un sesgo grave y destruye la estacionalidad\. Se propone usar interpolación lineal o ffill/bfill \(forward/backward fill\) para respetar la continuidad temporal de variables como \`wind\_speed\_10m\` o \`precipitation\`\.

 \- Justificación: La variable \`date\` debe ser convertida al tipo \`datetime\` explícito para extraer características vitales \(Mes, Hora\)\. Esto previene que el modelo trate el clima como eventos independientes y le permite entender el sesgo natural de las estaciones del año \(Ej: invierno = más lluvia y viento\)\.

   \- Justificación: Las ciudades en la columna \`Localidad\` no tienen una jerarquía matemática natural\. Usar Label Encoding \(1, 2, 3\) engañaría al algoritmo haciéndole creer que una ciudad "vale más" que otra\. Se usará One\-Hot Encoding \(variables dummy\) para mantener la equidad espacial\.

## 5\. Eliminación de Duplicados Temporales:

   \- Justificación: Filas repetidas para la misma hora y localidad sobre\-entrenarían el modelo en eventos específicos, generando sesgo de sobreajuste \(overfitting\)\.

# Proponer pasos a seguir considerando fases posteriores de acuerdo con los resultados obtenidos en las fases abordadas

## 6\. Propuesta de Pasos a Seguir \(Fases Posteriores\)

```

print("Dimensión original:", df.shape)

# RUTINA 1: Conversión y extracción de fechas 
# Convertir a datetime timezone-aware
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extraer características relevantes (ayuda a la precisión del modelo)
df['mes'] = df['date'].dt.month
df['hora'] = df['date'].dt.hour

# RUTINA 2: Eliminación de duplicados temporales y espaciales 
# Evitar sesgos por sobre-representación de una misma hora en una ciudad
df = df.drop_duplicates(subset=['date', 'Localidad'], keep='last')

# RUTINA 3: Imputación de Nulos (Interpolación) 
# Ordenamos por localidad y fecha para interpolar correctamente respetando el tiempo
df = df.sort_values(by=['Localidad', 'date']).reset_index(drop=True)

# Interpolación lineal para variables continuas para no perder la tendencia climática
cols_num = ['temperature_2m', 'relative_humidity_2m', 'wind_speed_10m', 'precipitation']
for col in cols_num:
    df[col] = df.groupby('Localidad')[col].transform(lambda x: x.interpolate(method='linear').bfill().ffill())

# RUTINA 4: Tratamiento de Outliers (Recorte por IQR) 
# Aplicamos clipping a wind_speed_10m para evitar sesgar el modelo por errores del sensor,
# pero reteniendo el tope superior porque los vientos fuertes son de alto interés.
Q1 = df['wind_speed_10m'].quantile(0.25)
Q3 = df['wind_speed_10m'].quantile(0.75)
IQR = Q3 - Q1
limite_superior = Q3 + 3 * IQR # Usamos 3 para conservar temporales reales, limitando solo valores imposibles

df['wind_speed_10m'] = np.where(df['wind_speed_10m'] > limite_superior, limite_superior, df['wind_speed_10m'])

# RUTINA 5: Codificación de Variables Categóricas sin jerarquía 
# One-Hot Encoding para la variable 'Localidad' (evita sesgo de jerarquía espacial)
df = pd.get_dummies(df, columns=['Localidad'], drop_first=False)

print("Dimensión después de limpieza:", df.shape)
print("Valores nulos restantes:\n", df.isnull().sum().sum())

# Mostrar un vistazo del dataset limpio
df.head()
```

```
# Detectar valores atípicos (outliers) en variables clave usando método IQR y Z-score
import numpy as np

# Seleccionamos variables numéricas relevantes
vars_numericas = ['wind_speed_10m','precipitation','relative_humidity_2m','temperature_2m','apparent_temperature','cloud_cover']

resultados = {}

# Método IQR por columna
for col in vars_numericas:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    low = q1 - 1.5*iqr
    high = q3 + 1.5*iqr
    mask_iqr = (df[col] < low) | (df[col] > high)
    resultados[col] = {
        'IQR_low': float(low),
        'IQR_high': float(high),
        'IQR_outliers': int(mask_iqr.sum())
    }

# Método Z-score (umbral 3)
from scipy import stats
for col in vars_numericas:
    col_vals = df[col].values
    # Evitar división por cero si std = 0
    if np.nanstd(col_vals) == 0:
        z_outliers = 0
    else:
        z = np.abs(stats.zscore(col_vals, nan_policy='omit'))
        z_outliers = int(np.sum(z > 3))
    resultados[col]['Z_outliers'] = z_outliers

# Mostrar resumen
import pandas as pd
resumen_outliers = pd.DataFrame(resultados).T
resumen_outliers = resumen_outliers[['IQR_low','IQR_high','IQR_outliers','Z_outliers']]
resumen_outliers
```
